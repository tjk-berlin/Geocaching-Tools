```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'chris_MRA.nii.gz'

img = nib.load(path)              # l√§dt Header + Affine + Proxy auf die Daten
data = img.get_fdata(dtype=np.float32)  # echte Daten als float32 in RAM (3D oder 4D)

print("Shape:", data.shape)       # z.B. (X, Y, Z) oder (X, Y, Z, T)
print("Affine:\n", img.affine)    # Weltkoordinaten-Transformation
print("Voxel sizes:", img.header.get_zooms())
```
